## Importare la libreria

```python
from intelligenzaartificiale import dataset as dt
from intelligenzaartificiale import statistica as st
```

## Importare o Caricare un Set di dati

```python
il_mio_dataset = dt.leggi_csv("exams.csv")
```

## Statistiche su Valori Nulli o Corrotti

```python
print(st.valori_nan(il_mio_dataset))
print("\n ----------------- \n")
print(st.percentuale_nan(il_mio_dataset))
```

## Statistiche su tutto il Set di dati

```python
print(st.statistiche(il_mio_dataset))
```

## Statistiche su specifica colonna

```python
#statistiche generiche su specifica colonna del dataset
print(st.statistiche_colonna(il_mio_dataset,"math score"))

print("\n ----------------- \n")

#media
print(st.media(il_mio_dataset,"math score"))

print("\n ----------------- \n")

#varianza
print(st.varianza_colonna(il_mio_dataset,"math score"))

print("\n ----------------- \n")

#quantili
print(st.quantile_25(il_mio_dataset,"math score"))
print(st.quantile_50(il_mio_dataset,"math score"))
print(st.quantile_75(il_mio_dataset,"math score"))

print("\n ----------------- \n")

#min e max
print(st.min(il_mio_dataset,"math score"))
print(st.max(il_mio_dataset,"math score"))
```

## Analizzare le correlazioni

```python
#Correlazione tra i campi del dataset
print(st.correlazione(il_mio_dataset))

print("\n ----------------- \n")

#correlazione tra una colonna target e un altra colonna
print(st.correlazione_radio(il_mio_dataset, "reading score" ,"math score"))

print("\n ----------------- \n")

#correlazione di Spearman tra una colonna target e un altra colonna
print(st.correlazione_spearman(il_mio_dataset, "reading score" ,"math score"))

print("\n ----------------- \n")

#correlazione di Pearson tra una colonna target e un altra colonna
print(st.correlazione_pearson(il_mio_dataset, "reading score" ,"math score"))

print("\n ----------------- \n")

#classifica correlazione tra una colonna target e un altra colonna
print(st.classifica_correlazione_colonna(il_mio_dataset, "math score"))
```
